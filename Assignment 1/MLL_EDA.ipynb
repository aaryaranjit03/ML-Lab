```python
unemploy_df = pd.read_csv('/content/drive/MyDrive/Datasets/clean_unemployment.csv')
```

```python
unemploy_df
```

```python
unemploy_df.describe()
```

```python
unemploy_df.info()
```

```python
unemploy_df.mean().plot(xlabel='Year', ylabel='Unemployment Rate', xticks=np.arange(0,12), grid=True, figsize=(10,5))
```

```python
unemploy_df[unemploy_df['continent']=='Asia'].mean().plot()
unemploy_df[unemploy_df['continent']=='Europe'].mean().plot()
unemploy_df[unemploy_df['continent']=='Africa'].mean().plot()
unemploy_df[unemploy_df['continent']=='Oceania'].mean().plot()
unemploy_df[unemploy_df['continent']=='North America'].mean().plot()
unemploy_df[unemploy_df['continent']=='South America'].mean().plot(figsize=(12,6),xticks=np.arange(0,12))
plt.legend(['Asia','Europe','Africa','Oceania','North America','South America'], loc='upper right', ncols=2, fontsize=8)
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('Yearly Mean Unemployment Rate by Continent')
```
